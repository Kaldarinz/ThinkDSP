```python
import numpy as np
import thinkdsp
from matplotlib import pyplot as plt
```

```python
wave1 = thinkdsp.SquareSignal().make_wave()
wave2 = thinkdsp.SquareSignal().make_wave()
wave1.segment(duration=.01).plot()
```

```python
conv = np.convolve(wave1.ys, wave2.ys, mode='same')
plt.plot(conv)
plt.gca().set_xlim()
```

```python
mult = np.fft.irfft(wave1.make_spectrum().hs**2)
plt.plot(mult)
plt.gca().set_xlim()
```
