```python
import numpy as np
```

```python
def fft(wave: np.ndarray) -> np.ndarray:

    # Check base condition
    if wave.size == 1:
        return wave
    # Split wave to arrays of even and odd elements
    evens = wave[1::2]
    odds = wave[::2]
    # Recursively calc ft for even and odd arrays
    fft_evens = np.tile(fft(evens), 2)
    fft_odds = np.tile(fft(odds), 2)
    result = fft_evens + np.exp(-2j*np.pi*np.arange(len(wave))/wave.size)*fft_odds
    return result
```

```python
import thinkdsp
from matplotlib import pyplot as plt
```

```python
freq = 20
framerate = 2**15
wave = thinkdsp.SquareSignal(freq=freq).make_wave(duration=1, framerate=framerate)
plt.plot(wave.ys)
```

```python
fft_wave = fft(wave.ys[:framerate])
real_fft = np.abs(fft_wave)[:fft_wave.size//2]
plt.plot(real_fft)
plt.gca().set_xlim((0,200))
real_fft.argmax()
```
